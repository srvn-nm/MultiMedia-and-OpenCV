## Imports

```python
# %pip install opencv-python
# %pip install matplotlib
# %pip install numpy
# %pip install --upgrade pip

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Image Loading and Visualization

```python
#imread
# Load the image
image_path = 'image/wikipedia_gamma_correction.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for Matplotlib

#visualization with matplotlib
# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
```

## Gamma Correction

```python
#implement gamma correction and save it to the new LUT

gamma = 2.2  # changing this value cause different results
inv_gamma = 1.0 / gamma
```

## LookUPTable mapping

```python
# Create a lookup table mapping pixel values [0, 255] to their gamma-corrected values
LUT = np.array([((i / 255.0) ** inv_gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")

#now map your image with its new LUT which contains correct contrast
gamma_corrected_image = cv2.LUT(image, LUT)
```

## Saving and Visualization

```python
#imsave
# Save the gamma-corrected image
cv2.imwrite('image/gamma_corrected_image.jpg', cv2.cvtColor(gamma_corrected_image, cv2.COLOR_RGB2BGR))

#visualization with matplotlib
plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected_image)
plt.title('Gamma Corrected Image')
plt.axis('off')

plt.show()
```
