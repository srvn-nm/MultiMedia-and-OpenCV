## Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Image Loading and Visualization

```python
# Function to display an image
def display_image(img, title='Image', cmap_type='gray'):
    plt.figure(figsize=(8, 8))
    plt.imshow(img, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Function to display histogram
def display_histogram(img, title='Histogram'):
    plt.figure(figsize=(8, 6))
    plt.hist(img.ravel(), bins=256, range=[0,256])
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
#imread, image: Unequalized_Hawkes_Bay_NZ.jpg
# Load the image
img_path = 'image/Unequalized_Hawkes_Bay_NZ.jpg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

#visualization with matplotlib 
# Show the original image
display_image(img, title='Original Image')
```

```python
#show the histogram plot of original image
display_histogram(img, title='Histogram of Original Image')
```

## Histogram Equalization

```python
# equalize the histogram of image
equalized_img = cv2.equalizeHist(img)
#show the histogram plot of original image
display_image(equalized_img, title='Equalized Image')
# Show the histogram plot of equalized image
display_histogram(equalized_img, title='Histogram of Equalized Image')

#output visualization
```

## Clahe(adaptive histogram equalization)

first import clache image and analyse its histogram plot

```python
#imread, image: clahe.jpg
# Load the CLAHE image
clahe_img_path = 'image/clahe.jpg'
clahe_img = cv2.imread(clahe_img_path, cv2.IMREAD_GRAYSCALE)

#visualization with matplotlib 
# Show the CLAHE image
display_image(clahe_img, title='CLAHE Image')
#show the histogram plot of original image
display_histogram(clahe_img, title='Histogram of CLAHE Image')
```

now do the clahe operation and global histogram equalization on clahe's images and analyse the outputs

```python
#global histogram equalization operation
global_eq_clahe_img = cv2.equalizeHist(clahe_img)
# Show the result of global histogram equalization on CLAHE image
display_image(global_eq_clahe_img, title='Global Histogram Equalized CLAHE Image')
# Show the histogram plot of global histogram equalized CLAHE image
display_histogram(global_eq_clahe_img, title='Histogram of Global Histogram Equalized CLAHE Image')

#clahe adaptive histogram equalization operation
# Apply CLAHE to the CLAHE image
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_result_img = clahe.apply(clahe_img)
#output visualization
```

```python
#show the histogram plot of clahe operation result
display_image(clahe_result_img, title='CLAHE Operation Result')
#show the histogram plot of global histogram equalization operation result
display_histogram(clahe_result_img, title='Histogram of CLAHE Operation Result')
```
